```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
btc = pd.read_csv('data/btc.csv', index_col='Date', parse_dates=True)
fng = pd.read_csv('data/feargreedindex.csv', index_col='date', parse_dates=True)
```

```python
btc
```

```python
fng
```

```python
df = pd.merge(btc.Open, fng.fng, left_index=True, right_index=True)
```

```python
df
```

```python
color_map = plt.cm.seismic_r
plt.figure(figsize=(20,12))
plt.scatter(df.index, df.Open, c=df.fng, cmap=color_map)
plt.plot(df.Open, c='Grey')
#plt.yscale('log')
```
