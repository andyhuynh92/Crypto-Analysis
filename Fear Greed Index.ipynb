```python
import pandas as pd
import datetime
import requests
```

```python
response = requests.get("https://api.alternative.me/fng/?limit=0&format=csv&date_format=cn")
```

```python
tmp = response.text
```

```python
rows = tmp.split('\n')
```

```python
rows = rows[3:-6]
```

```python
date = [datetime.date(*[int(i) for i in row.split(',')[0].split('-')]) for row in rows]
fng = [row.split(',')[1] for row in rows]
classification = [row.split(',')[2] for row in rows]

pd.DataFrame({'date': date, 'fng': fng, 'classification': classification}).to_csv('data/feargreedindex.csv', index=False)
```
