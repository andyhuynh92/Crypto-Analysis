```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
btc = pd.read_csv('data/btcall.csv', index_col='Date', parse_dates=True).sort_values('Date').apply(lambda x: x.str.replace(',', ''))
```

```python
halvingdates = [datetime.date(2012, 11, 28), datetime.date(2016, 7, 9), datetime.date(2020, 5, 11)]
```

```python
half1 = btc[halvingdates[0]:halvingdates[1]].reset_index().Price.astype('float')
half2 = btc[halvingdates[1]:halvingdates[2]].reset_index().Price.astype('float')
half3 = btc[halvingdates[2]:].reset_index().Price.astype('float')
```

```python
df = pd.concat([half1 / half1[0], half2 / half2[0], half3 / half3[0]], axis=1)
df.columns = ['First Halvening', 'Second Halvening', 'Third Halvening']
```

```python
plt.figure(figsize=(40, 24))
plt.plot(df)
plt.legend(df.columns)
plt.title('Halvening Cycles')
```

```python
plt.figure(figsize=(40, 24))
plt.plot(df)
plt.legend(df.columns)
plt.yscale('log')
plt.title('Halvening Cycles')
```
